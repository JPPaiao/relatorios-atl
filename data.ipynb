```python
def get_units(unidade):
  # req = requests.get(f"https://comex.io/v1/conteiner/tracking?api_key=2a7959e9189097e782738190d910abad&conteiner={unidade}")
  req = requests.get(f"https://api.logcomex.com.br/v2/get_conteiner_tracking?api_key={LOGCOMEX_API}&conteiner={unidade}")
  return json.loads(req.content)

def get_cnpj_unit(unidade):
  datas = get_units(unidade)

  # if datas['response_status'] != "ok":
  #   while datas['response_status'] != "ok":
  if datas['status'] != "ok":
    while datas['status'] != "ok":
      # time.sleep(50)
      datas = get_units(unidade)
    # print('erro')
    
    return datas['data']

  return datas['data']


```

```python
# Suponha que get_cnpj_unit retorna uma lista de dicionários, que você converte em DataFrame
datas = pd.DataFrame(get_cnpj_unit('TCLU1310328'))
# datas = pd.DataFrame(get_cnpj_unit('PCIU9310218'))

# Converte a coluna 'data_operacao' para datetime
datas['data_operacao'] = pd.to_datetime(datas['data_operacao'], format='%Y-%m-%d')

# Filtra os dados onde o CNPJ não está vazio
dados_validos = datas[datas['importador_cnpj'] != '']

# Encontra o mês e o ano mais recentes
dados_validos = dados_validos[dados_validos['data_operacao'].dt.year == dados_validos['data_operacao'].dt.year.max()]
dados_validos = dados_validos[dados_validos['data_operacao'].dt.month == dados_validos['data_operacao'].dt.month.max()]

# Ordena o DataFrame, priorizando o 'HBL' no tipo_conhecimento
dados_ordenados = dados_validos.sort_values(by='tipo_conhecimento', key=lambda x: x != 'HBL')

# Pega o CNPJ mais recente que atenda às condições
cnpj_mais_recente = dados_ordenados.iloc[0]['importador_cnpj'] if not dados_ordenados.empty else None

print(dados_ordenados.iloc[0]['tipo_conhecimento'])

print()
print(cnpj_mais_recente)

# Exibe as informações desejadas do DataFrame
for _, row in dados_ordenados.iterrows():
    print(row['tipo_conhecimento'], '-', row['importador_cnpj'], '-', row['data_operacao'])
```

```python
df_depot['remarks'] = df_depot['remarks'].fillna('')
# print(df_depot['remarks'])

df_cobranca = pd.DataFrame({
  "UNIDADE": df_depot["cntr"],
  "TIPO": df_depot['type'],
  "OWNER": df_depot['own'],
  "ENTRADA": df_depot['datein'],
  "CNPJ AGENDADO": df_depot["importer"],
  "CNPJ HBL": "",
  "TRANSPORTADORA": df_depot["carrier"],
  "CNPJ TRANSPORTADORA": df_depot["carrier_cnpj"],
  "VALORES": df_depot['price_use'],
  "OBS": "",
  "DATA. PAG": df_depot['payment'],
  "NF": df_depot['number nf'],
  "ISENTO": "",
  "V. ISENTO": ""
})

df_cobranca['NF'] = df_cobranca['NF'].fillna('')
df_cobranca['DATA. PAG'] = df_cobranca['DATA. PAG'].fillna('')

df_cobranca['CNPJ HBL'] = df_cobranca['CNPJ HBL'].astype("str")
df_cobranca['CNPJ AGENDADO'] = df_cobranca['CNPJ AGENDADO'].astype("str")
df_cobranca['CNPJ TRANSPORTADORA'] = df_cobranca['CNPJ TRANSPORTADORA'].astype("str")

df_cobranca['CNPJ HBL'] = df_cobranca['CNPJ HBL'].fillna('')
df_cobranca = df_cobranca.dropna(subset=['CNPJ HBL'])

df_cobranca = df_cobranca.dropna(subset=['CNPJ AGENDADO'])
df_cobranca = df_cobranca.dropna(subset=['ENTRADA'])
df_cobranca = df_cobranca.drop_duplicates(subset=['UNIDADE'])

for unidade in tqdm(df_cobranca['UNIDADE'], desc="Processando unidades"):
  isento = df_depot.loc[df_depot['cntr'] == unidade, 'remarks'].values[0]
  
  if isento != '':
    isento_separado = isento.split(' - ')

    # for item in isento_separado:
    #   df_cobranca.loc[df_cobranca['UNIDADE'] == unidade, 'ISENTO'] = isento_separado[0]
    #   df_cobranca.loc[df_cobranca['UNIDADE'] == unidade, 'V. ISENTO'] = isento_separado[1]

  df_get_uni = pd.DataFrame(get_cnpj_unit(unidade))

  dados_validos = df_get_uni[df_get_uni['importador_cnpj'] != '']
  dados_validos = df_get_uni.dropna(subset=['importador_cnpj'])

  dados_validos['data_operacao'] = pd.to_datetime(dados_validos['data_operacao'], format='%Y-%m-%d')

  dados_validos = dados_validos[dados_validos['data_operacao'].dt.year == dados_validos['data_operacao'].dt.year.max()]
  dados_validos = dados_validos[dados_validos['data_operacao'].dt.month == dados_validos['data_operacao'].dt.month.max()]

  dados_ordenados = dados_validos.sort_values(by='tipo_conhecimento', key=lambda x: x != 'HBL')

  cnpj_mais_recente = dados_ordenados.iloc[0]['importador_cnpj'] if not dados_ordenados.empty else ''
  
  if dados_ordenados.empty:
    df_cobranca.loc[df_cobranca["UNIDADE"] == unidade, 'CNPJ HBL'] = cnpj_mais_recente
    continue

  if dados_ordenados.iloc[0]['tipo_conhecimento'] != 'HBL':
    df_cobranca.loc[df_cobranca["UNIDADE"] == unidade, 'OBS'] = f'SEM HBL/AGENDADO NO {dados_ordenados.iloc[0]["tipo_conhecimento"]}'

  df_cobranca.loc[df_cobranca["UNIDADE"] == unidade, 'CNPJ HBL'] = cnpj_mais_recente

df_cobranca['CNPJ AGENDADO'] = df_cobranca['CNPJ AGENDADO'].fillna('')
df_cobranca = df_cobranca.drop_duplicates(subset=['UNIDADE'])

df_cobranca.to_excel('cobrana-21-11.xlsx', index=False)
df_cobranca
```
